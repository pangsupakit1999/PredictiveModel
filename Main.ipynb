```python
#best_combination = None
#best_score = float('inf')
#test_sizes = [0.5, 0.6, 0.7]
#optimizer_types = ['Adam', 'SGD', 'RMSprop']
#hidden_layer_numbers = [8, 32, 64]
#hidden_sizes = [8, 16, 32]
#learning_rates = [0.001, 0.01, 0.1]
#epochs = [200, 500, 1000]
#batch_sizes = [32, 64, 128]
#activations = ['relu', 'tanh', 'sigmoid']
#loss_functions = ['MSE', 'MAE', 'Huber']

ann_model = Model.ANN()
best_combination = None
best_score = float('inf')
test_sizes = [0.7]
optimizer_types = ['RMSprop']
hidden_layer_numbers = [2]
hidden_sizes = [4]
learning_rates = [0.001]
epochs = [100]
batch_sizes = [256, 512]
activations = ['sigmoid']
loss_functions = ['MSE', 'MAE']

total_combinations = (
    len(test_sizes) *
    len(optimizer_types) *
    len(hidden_layer_numbers) *
    len(hidden_sizes) *
    len(learning_rates) *
    len(epochs) *
    len(batch_sizes) *
    len(activations) *
    len(loss_functions)
)

completed_combinations = 0

for test_size in test_sizes:
    for optimizer_type in optimizer_types:
        for hidden_layer_number in hidden_layer_numbers:
            for hidden_size in hidden_sizes:
                for lr in learning_rates:
                    for ep in epochs:
                        for batch_size in batch_sizes:
                            for activation in activations:
                                for loss in loss_functions:
                                    completed_combinations += 1
                                    progress_percentage = (completed_combinations / total_combinations) * 100
                                    
                                    model, r2_set = ann_model.train(target_set=target_set, 
                                                                 feature_set=feature_set, 
                                                                 nums_outputs=nums_outputs,
                                                                 test_size=test_size, 
                                                                 optimizer_type=optimizer_type,
                                                                 hidden_layer_number=hidden_layer_number, 
                                                                 hidden_size=hidden_size, 
                                                                 lr=lr, 
                                                                 ep=ep, 
                                                                 batch_size=batch_size,
                                                                 activation=activation, 
                                                                 loss = loss, 
                                                                 metrics='MAE')
                                    
                                    current_score = sum(r2_set)/2
                                    if current_score > best_score: 
                                        best_score = current_score
                                        best_combination = (test_size, optimizer_type, hidden_layer_number,
                                                            hidden_size, lr, ep, batch_size, activation, loss)
                                        
                                    print(f'Progress: {progress_percentage:.2f}% | '
                                          f'Test Size: {test_size}, Optimizer: {optimizer_type}, '
                                          f'Hidden Layers: {hidden_layer_number}, Hidden Size: {hidden_size}, '
                                          f'Learning Rate: {lr}, Epochs: {ep}, Batch Size: {batch_size}, '
                                          f'Activation: {activation}, Loss: {loss}')

print(f'Best Combination: {best_combination}')
print(f'Best Score: {best_score}')
```
